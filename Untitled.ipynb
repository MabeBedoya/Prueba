<H1> TAREAS CD993

```python
Tarea=Tarea93_MAX[["LOCAL","TAREA","CODIGOARTICULO","VALORUNIDADMANEJO","UBICACION"]]
Tarea.rename({"TAREA":"ID_PALLET","CODIGOARTICULO":"CODIGO_BARRAS","VALORUNIDADMANEJO":"UNIDAD MANEJO"}, axis=1,inplace=True)
Tarea["NAVE"]=scipy.int64(Tarea.UBICACION/1000000000)

# ORDEN DE PICKING
Tarea["valor"]=1
Tarea['ORDEN']=Tarea.groupby('ID_PALLET')['valor'].cumsum()
Tarea.drop(['UBICACION', 'valor'], axis='columns', inplace=True)


Tarea=Tarea.sort_values(by='LOCAL', ascending=False)
Tarea.head(3)
```
