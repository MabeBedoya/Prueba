```python
# Transformar Fechas 1900 y 3000 a fecha actual (siguen trabajando)
Fecha["nom_fec"]=np.where((Fecha["nom_fec"]=="1900-01-01")|(Fecha["nom_fec"]=="3009-06-01")|(Fecha["nom_fec"]=="3006-09-30")|(Fecha["nom_fec"]=="3004-07-05"), Fechactual, Fecha["nom_fec"] )
```

```python
# Transformar fecha datatime
Formato="%Y-%m-%d"
Fecha["nom_fec"]=pd.to_datetime(Fecha["nom_fec"], format=Formato)
```

## Evaluación

```python
%%storage read --object "gs://bbedoya/DatosColaboradores/evaluacionescf.csv" --variable evaluacionescf
```

```python
EvaluacionCF=pd.read_csv(BytesIO(evaluacionescf),sep=";")
```

```python
# Evaluaciones datos útiles
Evaluacion=EvaluacionCF[['ID_FECHA_INI', 'ID_FECHA_FIN', 'ID_EMPLEADO', 'NOM_OCUP',"NOM_ZONA_LUGAR",'NOM_FACTOR','puntaje_factor_maximo',
                                     'puntaje_factor_obtenido',"NOM_PERIODO"]]
```

```python
# Join Fecha Inicio
Evaluacion1=pd.merge(left=Evaluacion, right=Fecha,
                       how="inner",left_on="ID_FECHA_INI",right_on="id_fecha")
Evaluacion1=Evaluacion1.rename({'nom_fec':'Fecha_Inicia'}, axis=1)

#Join Fecha Fin
Evaluacion2=pd.merge(left=Evaluacion1, right=Fecha,
                       how="inner",left_on="ID_FECHA_FIN",right_on="id_fecha")
# Reombrando columnas
Evaluacion2=Evaluacion2.rename({'nom_fec':'Fecha_Fin'}, axis=1)
Evaluacion2.head()

#Extracción de columnas "útiles"
EvaluacionFinal=Evaluacion2[['Fecha_Inicia','Fecha_Fin','ID_EMPLEADO', "NOM_ZONA_LUGAR",'NOM_OCUP','NOM_FACTOR','puntaje_factor_maximo','puntaje_factor_obtenido','NOM_PERIODO']]
```

```python
# Dias trabajados
EvaluacionFinal["Dias_Trabajados"]=EvaluacionFinal["Fecha_Fin"]-EvaluacionFinal["Fecha_Inicia"]

# Cambia el tipo de dato, de datatime a entero
EvaluacionFinal["Dias_Trabajados"]=EvaluacionFinal["Dias_Trabajados"].dt.days
```

```python
# Quitar espacios en blanco de las ocupaciones
Sinespac=EvaluacionFinal["NOM_OCUP"].map(lambda x: x.replace(' ', ''))
EvaluacionFinal["NOM_OCUP"]=Sinespac
```

```python
# Agrupación
EvaAgrup=EvaluacionFinal.groupby("NOM_OCUP")
EvaAgrup.size()
```

```python
EvaluacionFinal["NOM_OCUP"]
```

```python
# Ocupación: CAJERO * - CAJEROS 
Cajero1=EvaAgrup.get_group("CAJERO*")
Cajero2=EvaAgrup.get_group("CAJEROS")
CAJEV=pd.concat([Cajero1,Cajero2],axis=0) 
# Ocupación: EMPACADOR-EMPACADOR * 
Empacador=EvaAgrup.get_group("EMPACADOR")
Empacador1=EvaAgrup.get_group("EMPACADORES")
EMPEV=pd.concat([Empacador,Empacador1])
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

import plotly.plotly as py
from plotly.graph_objs import *
import plotly.tools as tls

from sklearn.decomposition import PCA as sk_pca
```

```python
X=IdFac.iloc[:,1:22].values
y=IdFac.iloc[:,0].values
X_stad=StandardScaler().fit_transform(X)
```

```python
acp=sk_pca(n_components=5)
Y=acp.fit_transform(X_stad)
```

```python
Y
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import numpy as np

result=[]
for name in ("setosa","versiolor","virginica"):
    result=Scatter (x=Y[y==name,0],y=Y[y==name,1],
                   mode="markers",name=name,
                   marker=Marker(size=15,line=Line(color="rgb{225,225,225,0.2}",width=0.5),opacity=0.75))
    results=append(result)
    
data=Data(results)
layout=Layout(showlegend=True, scene=Scene(xaxis=XAxis(title="CP1"),
                                          yaxis=YAxis(title="CP2")))
fig=Figure(data=data,layout=layout)
py.iplot(fig)
```
